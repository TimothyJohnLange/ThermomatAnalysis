```python
%pylab inline
from datahandling import alldatafiles, DataFile, file_parse
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages
```

```python
Files = alldatafiles()
```

```python
with PdfPages('compare_test.pdf') as pdf:
    for i in [[13,18],[26,35],[37,38],[5,21]]:
        fig = figure()
        for j in i:
            f = Files[j]
            sample_no = file_parse(f)
            time_data, conduct_data = DataFile(f).simple_data()
            
            my_label = 'Sample Number ' + sample_no
            plot(time_data, conduct_data, label=my_label)
        
        legend()
```
