```python
%pylab inline
from glob import glob
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages
```

```python
Files = glob('*.txt')
```

```python
sample_number_all = []
stability_time_all = []

with PdfPages('thermomat_results_plot.pdf') as pdf:
    for filename in Files:
        split_filename = filename.split(' ')
        sample_number = split_filename[1]

        data = pd.read_table(filename, skiprows = 4, sep=';')

        time_data = data['s'].values
        time_data = time_data/60
        conduct_data = data[data.columns[1]].values
        
        done = False
        failure_crit = 50 # microSiemens/ccm
        for i, v in enumerate(conduct_data):
            if v >= failure_crit and done == False:
                stability_time = time_data[i]
                stability_cond = v
                done = True
                
        if done == False:
            stability_time = 0.0
            stability_cond = 0.0
            
        sample_number_all.append(float(sample_number))
        stability_time_all.append(round(stability_time, 1))
        
        fig = figure()
        fig.suptitle('Sample Number ' + sample_number + ' with Stab. Time of ' + str(round(stability_time, 1)) + ' min')
        plot(time_data, conduct_data, label='Conductivity')
        plot(stability_time, stability_cond, '*')
        
        pdf.savefig(fig)

stab_time_data = {'Sample Number': sample_number_all,
                  'Stability Time (min)': stability_time_all}

stab_time_df = pd.DataFrame(data=stab_time_data)

stab_time_df = stab_time_df.sort('Sample Number', ascending=True)

# stab_time_df.to_csv('Stability Time Data.csv')
```
